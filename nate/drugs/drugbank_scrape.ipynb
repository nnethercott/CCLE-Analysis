# NOW FOR THE REAL(ISH) SCRAPE

```python
#gonna try this process out on a list of drugs we already know are in the database 
sample_drugs = ['Motesanib',
                 'Veliparib',
               'Zibotentan',
               'Amuvatinib']

search_url = 'https://go.drugbank.com/unearth/q?searcher=drugs&query='
```

```python
#functions for easy text extraction given the page exists 

#returns a partially filled, labelled row
def summary_row_extraction(summary_soup):
    col_names = [item.text for item in summary_soup.find_all("dt")]
    col = {item:[] for item in col_names}

    for i,desc in enumerate(summary_soup.find_all("dd")):
        col[col_names[i]] = desc
    
    return col


#just get the soup object 
def pharm_info_extraction(pharm_soup):
    return pharm_soup.find("dd")
```

```python
#okay generate all our drugs for the scrape 
drugs_modified = list(map(lambda x: ''.join(x.split(' ')[:-1]) if len(x.split(' '))>1 else x, drugs))
drugs_modified
```

```python
from tqdm import tqdm
import time

drug_df_prototype = []
#----------------------------------------------------------
#LOOP (should run in parallel for better performance but hey)

for i,name in enumerate(tqdm(drugs_modified)):
        
    r = requests.get(search_url+name)
    soup = BeautifulSoup(r.text, 'html.parser')
    
    #get main page info we're interested in
    content = soup.find("div", {'class':"card-content px-md-4 px-sm-2 pb-md-4 pb-sm-2"})
    
    #populate
    try:
        real_name = drugs[i]
        #real_name = name
        
        #info from first cell
        summary_soup = content.find("dl");  
        summary_info = summary_row_extraction(summary_soup)
        summary_info["NAME"] = real_name;
        
        #info from second 
        pharm_soup = content.find_all("dl")[2];
        pharm_info = pharm_info_extraction(pharm_soup)
        
        #combine our info 
        row_info = summary_info
        row_info['Indication'] = pharm_info
        
        #append
        drug_df_prototype.append(row_info)
        
    except:
        row_info = {"NAME":real_name}
        drug_df_prototype.append(row_info)
        
```

```python
drug_df = pd.DataFrame(drug_df_prototype)
drug_df.index = drug_df["NAME"]
drug_df.shape
```

```python
#process/trim the dataframe
keep = ['Summary', 'Background', 'Weight', 'DrugBank Accession Number', 'Groups', 'Indication']
cols_to_drop = list(set(list(drug_df.keys()))-set(keep))
drug_df = drug_df.drop(cols_to_drop, axis=1)
drug_df
```

```python
drug_df = drug_df.applymap(lambda x: x.text if type(x)!=float else x)
drug_df['Weight'] = drug_df['Weight'].apply(lambda x: float(x.split(' ')[1]) if type(x)!=float else x)
```

```python
drug_df["Indication"] = drug_df["Indication"].apply(lambda x: "Not Available" if type(x)!=float and "Not Available" in x else x)
```
